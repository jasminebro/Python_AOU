# Filtering data

```python
#filter example 

# declare a dictionary
record = { 
  
 'Name' : ['Ankit', 'Swapnil', 'Aishwarya', 
          'Priyanka', 'Shivangi', 'Shaurya' ],
    
 'Age' : [22, 20, 21, 19, 18, 22], 
    
 'Degree' : ['Math', 'Commerce', 'Science', 
            'Math', 'Math', 'Science'], 
    
 'Grad Year' : [1990, 1990, 1996, 1975, 1970, 1980] } 
    
# create a dataframe 
dataframe = pd.DataFrame(record,
                         columns = ['Name', 'Age', 
                                    'Degree', 'Grad Year']) 
# show the Dataframe
print("Given Dataframe :\n", dataframe)
```

```python
# selecting rows based on condition
rslt_df = dataframe[dataframe['Grad Year'] > 1970]
print('\nResult dataframe :\n', rslt_df)
```

```python
# selecting rows based on condition

#create a list of items we are looking for
Degree = ['Science', 'Commerce']

#select based on conditions in our list
rslt_df = dataframe[dataframe['Degree'].isin(Degree)]

print('\nResult dataframe :\n',rslt_df)
```

# Merging and joining data

```python
#merge: for combining data on common columns or indices
#dataframe.merge(right, how, on, left_on, right_on, left_index, right_index, sort, suffixes, copy, indicator, validate) 

data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

data2 = {
  "name": ["Sally", "Peter", "Micky"],
  "age": [77, 44, 22]
}
```

```python
df_1 = pd.DataFrame(data1)
df_2 = pd.DataFrame(data2)

df_1
```

```python
df_2
```

```python
newdf = df_1.merge(df_2, how='outer') 

newdf
```

```python
newdf2 = df_1.merge(df_2, how='left') 

newdf2
```

```python
newdf3 = df_1.merge(df_2, how='right') 

newdf3
```

```python
newdf4 = df_1.merge(df_2, how='inner') 

newdf4
```

```python
newdf5 = df_1.merge(df_2, how='cross') 

newdf5
```

```python
#join: for combining data on a key column or an index
#dataframe.join(other, on, how, lsuffix, rsuffix, sort)

data56 = {
  "name": ["Rural", "City", "Metro"],
  "zone": [50, 40, 30]
}

data57 = {
  "qualified": [True, False, False]
}
```
