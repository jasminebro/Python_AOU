## Example of statistical analysis for this DataFrame

```python
# Subset data
male = df.query('sex == "male"')['fasting_sugar']
female = df.query('sex == "female"')['fasting_sugar']
```

```python
df.groupby('sex').describe()
```

```python
#t-test

test = stats.ttest_ind(male, female, 
                      equal_var=True)

display(test)
```

# Seaborn

Matplotlib is used to plot graphs using Pandas and NumPy. Seaborn is an extended version of Matplotlib that uses Pandas and Numpy as well. 

Matplotlib works efficiently with DataFrames and arrays. It treats figures and axes as objects. Therefore, plot()-like methods can work without parameters.

Seaborn is more functional and organized than Matplotlib and treats the whole dataset as a single unit.  Parameters are required while calling methods like plot().

```python
#boxplot with seaborn

import seaborn as sns

sns.boxplot(x='sex', y='fasting_sugar', data=df)
```

```python
#boxplot without seaborn

df.boxplot(column=['fasting_sugar'], by=['sex'])
```

```python
#saving a seaborn plot

fasting_sugar_by_sex_plot = df.boxplot(column=['fasting_sugar'], by=['sex'])
fig = fasting_sugar_by_sex_plot.get_figure()
fig.savefig("fasting_sugar_by_sex_plot.png")
```
