# Aggregation

Aggregation in Pandas provides various functions that perform a mathematical or logical operation and returns a summary. 

Aggregation can be used to get a summary of columns like the sum, minimum, maximum, etc. 

Some functions used in the aggregation are:

sum() - returns the sum of the column\
min() - returns the min of the column\
max() - returns the max of the column\
mean() - returns the mean of the column\
size() - returns column size\
describe() - returns descriptive statistics\
first() - computes first values in group note: only supports a DatetimeIndex index\
last() - computes last values in group  note: only supports a DatetimeIndex index\
count() - counts column values\
std() - returns the standard deviation of a column\
var() - computes the variance of a column\
sem() - computes the standard error of the mean of a column

```python
df.sum() #calculates the sum of every column
```

```python
df.describe() #descriptive statistics
```

```python
#you can use the agg() function to calculate the sum, min, and max of each column 

df.agg(['sum', 'min', 'max'])
```

```python
df.count() #counts column values
```

```python
df.std() #calculates the standard deviation
```

```python
df.var() #calculates the variance
```

```python
df.sem() #calculates the standard error of the mean
```

# Grouping 

Grouping is used to group data using some criteria from our dataset. It is used as split-apply-combine strategy.
1. Split the data into groups
2. Apply a function to each group
3. Combine the results

```python
df.groupby("fasting_sugar") #an object
```

```python
#view groups visually

a = df.groupby("fasting_sugar")
a.first()
```

```python
# gives the count of the number of groups 

df.groupby("sex").ngroups
```

```python
#group by fasting sugar and get the sum of all the columns

df.groupby("fasting_sugar").sum()
```

```python
#group by sex and get the sum of all the columns

df.groupby("sex").sum() 
```

```python
#group by two variables and get the min

#Note: In the above two cells, the results are not being saved to
#a new DataFrame but are being viewed.
#You can save any changes or operations to a new DataFrame per below

df2 = df.groupby(["sex", "age_at_diagnosis"]).agg("min")
```

```python
df2
```

```python
#group by sex and keep only the female group

df3 = df.groupby("sex").get_group("female")
```

```python
df3
```

```python
# return the first two records of each group

df4 = df.groupby(["sex"]).head(2)
```
